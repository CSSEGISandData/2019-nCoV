```python
import folium
from folium import plugins
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import os
from IPython.display import display
import math
```

```python
for dirname, _, filenames in os.walk('./archived_data/archived_time_series'):
    for file in filenames:
        if file.endswith('Confirmed.csv'):
            confirmed_data = pd.read_csv(os.path.join(dirname,file))
        elif file.endswith('Deaths.csv'):
            death_data = pd.read_csv(os.path.join(dirname,file))
        elif file.endswith('Recovered.csv'):
            recovered_data = pd.read_csv(os.path.join(dirname,file))
```

```python
confirmed_data['Log Suspected'] = confirmed_data.iloc[:,-1].apply(lambda s: math.log(s))
```

```python
import plotly.graph_objects as go

fig = go.Figure()
```

```python
dates = confirmed_data.iloc[:,4:].columns
```

```python
for date in dates:
    fig.add_trace(go.Densitymapbox(lat=confirmed_data['Lat'],lon=confirmed_data['Long'], 
                                   z=np.log(confirmed_data[date]), radius=10, opacity=1))
    fig.update_layout(mapbox_style="stamen-terrain", mapbox_center_lon=180)
```

```python
steps = []
for i in range(len(fig.data)):
    step = dict(
        method="restyle",
        args=["visible", [True] * len(fig.data)],
        name=str(confirmed_data.filter(regex='/*2').columns.to_list())

    )
    step["args"][1][i] = True
    steps.append(step)
```

```python
sliders = [dict(
    active=0,
    currentvalue={"prefix": "Frequency: "},
    pad={"b": 10, "t": 50},
    steps=steps
)]
```

```python
updatemenus = [
    {
        'buttons': [
            {
                'args': [[None], {'frame': {'duration': 500, 'redraw': False},
                         'fromcurrent': True, 'transition': {'duration': 300, 'easing': 'quadratic-in-out'}}],
                'label': 'Play',
                'method': 'animate'
            },
            {
                'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate',
                'transition': {'duration': 0}}],
                'label': 'Pause',
                'method': 'animate'
            }
        ],
        'direction': 'down',
        'pad': {'r': 100, 't': 100},
        'showactive': False,
        'type': 'buttons',
        'x': 0.1,
        'xanchor': 'right',
        'y': 0,
        'yanchor': 'top'
    }
]
```

```python
fig.update_layout(
    sliders=sliders,
    updatemenus=updatemenus)
fig.show()
```

```python
confirmed_copy = confirmed_data.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name="Date", value_name="Value").copy()
confirmed_copy.drop('Province/State', axis=1, inplace=True)
```
